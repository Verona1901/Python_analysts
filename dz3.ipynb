##Задача 1
Скачать данные по ссылке https://www.kaggle.com/datasets/astronautelvis/kc-house-data
Считать данные с помощью pandas
Вывести на экран первые 5 строк

```python
import pandas as pd


df = pd.read_csv('kc_final.csv')
df.head(5)
```

###1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
Если реновации дома не было, то в новом признаке поставьте 0

- yr_renovated - Год ремонта дома
- yr_built - Год постройки дома

```python
df['delta_renovated'] = df['yr_renovated'] - df['yr_built']
df['delta_renovated'] = df['delta_renovated'].apply(lambda x: x if x > 0 else 0 )
df.head()
```

###1.3 Создайте признаки года продажи, месяца продажи

-year_sale

-month_sale

```python
df['year_sale'] = df['date'].map(lambda x: str(x)[:4])
df['month_sale'] = df['date'].map(lambda x: str(x)[5:7])
df.head()
```

```python
# 2 способ

from datetime import datetime

date_str = '2014-10-13'
date_obj = datetime.strptime(date_str, '%Y-%m-%d')
year = date_obj.year
month = date_obj.month

# Преобразование столбца с датами в формат даты Python
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')

# Создание признака "года продажи"
df['year_sale'] = df['date'].apply(lambda x: x.year)

# Создание признака "месяца продажи"
df['month_sale'] = df['date'].apply(lambda x: x.month)

df.head()
```

```python
df[['date','year_sale','month_sale']].head()
```

###1.4 Удалите признаки date, zipcode, lat, long

```python
df.drop(columns = ['date','zipcode','lat', 'long'], inplace = True)
df.head()
```

###3.1 Найдите среднюю стоимость домов в зависимости от количества спален
Отсортируйте от меньшей стоимости к большей

- bedrooms - Кол-во спален

```python
df.groupby('bedrooms').agg({'price': 'mean'}).sort_values('price')
```

###3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

- condition - Значение от 1 до 5 насколько хорошее состояние

```python
df.groupby('condition')['price'].describe()[['min', 'mean', 'max']]
```

###3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

- waterfront - Есть набержная или нет
- view - Значение от 0 до 4 насколько хороший вид

```python
pd.pivot_table(df, index='waterfront', columns = 'view', values = 'id', aggfunc='count')
```

###3.4 Каких домов в зависимости от этажности и количества спален больше?

- floors - Кол-во этажей
- bedrooms - Кол-во спален


```python
pd.crosstab(index = df['floors'], columns = df['bedrooms'])
```

###3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

- condition - Значение от 1 до 5 насколько хорошее состояние
- grade - Значение от 1 до 13, где 1-3 плохая конструкция здания и дизайн, 7 - средний уровень конструкции и дизайна, 11-13 - высокое качество конструкции и дизайна

```python
pd.pivot_table(df, index='grade', columns ='condition', values = 'price', aggfunc='median')
```
