```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np



df = pd.read_csv('kc_house_data.csv')
df.head()
```

###1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize=(6,6))
plt.hist(df['price'],bins=100)
plt.title('Распределение стоимости')
plt.xlabel('Цена')
plt.ylabel('Количество');
```

```python
plt.figure(figsize=(12, 4))

plt.hist(
    df['price'],
    bins='auto',
    log=True
)
plt.title("Распределение стоимости")
plt.xlabel("Цена");
```

```python
df['price'].describe()
```

Половина домов находится в ценовом диапазоне от 320 до 645 тыс. Чем выше стоимость, тем меньше домов. Имеются отдельные дома с существенно более высокой ценой.

###1.3 Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(6, 6))
sns.histplot(df['sqft_living'], bins=20)
plt.title('Распределение жилой площади')
plt.xlabel('Кв. метры жилые')
plt.ylabel('Количество');
```

Половина домов сосредоточена в узком диапазоне (от 1427 до 2550 кв.м). Заметно снижение количества домов с увеличением жилой площади. Так же имеются дома, с заметно большей площадью.

###1.4 Изучите распределение года постройки

##Задача 3
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

Таким образом, можно сделать вывод, что стоимость недвижимости сильно зависит от жилой площади (sqft_living) и качества конструкции и дизайна (grade), а так же наблюдается корреляция от количества ванных (bathrooms) и кол-ва площади выше земли (sqft_above). Незначительно так же на цену влияет вид на набережную, вид из окна, этажность дома, количество спален, наличие подвала.

```python
sns.jointplot(x=df['sqft_living'], y=df['price'], kind='reg')

plt.xlabel("Жилая площадь, кв. м")
plt.ylabel("Цена дома");
```

По графику видна зависимость стоимости недвижимости от размера его жилой площади.

```python
sns.jointplot(x=df['grade'], y=df['price'], kind='reg')

plt.xlabel("Дизайн")
plt.ylabel("Цена дома");
```

На графике видно, что чем выше качество конструкции и дизайна, тем более высокая цена на недвижимость.

```python
sns.jointplot(x=df['bathrooms'], y=df['price'], kind='reg')

plt.xlabel("Количество ванных комнат")
plt.ylabel("Цена дома");
```

```python
sns.boxplot(x=df['bathrooms'], y=df['price'])

plt.title('Взаимосвязь стоимости дома и количества ванных комнат')
plt.xlabel("Количество ванных комнат")
plt.ylabel("Цена дома");
```

По графику виден рост цены с увеличением количества ванных комнат. (Данная ситуация наблюдается практически по всем домам, за исключением некоторых выбросов)

```python
sns.jointplot(x=df['sqft_above'], y=df['price'], kind='reg')

plt.xlabel("Площадь чердака")
plt.ylabel("Цена дома");
```

Заметно влияние на рост цены наличие и размер чердака.

```python
sns.boxplot(x=df['floors'], y=df['price'])

plt.title('Взаимосвязь стоимости дома и количества этажей')
plt.xlabel("Количество этажей")
plt.ylabel("Цена дома");
```

По данному графику, можно сделать вывод, что наиболее высокая цена наблюдается у домов с 2,5 этажами.

```python
sns.boxplot(x=df['waterfront'], y=df['price'])

plt.title('Взаимосвязь стоимости дома и наличия вида на набережную')
plt.xlabel("Наличие набережной")
plt.ylabel("Цена дома");
```

Наблюдается более высокая цена на недвижимость при наличии вида на набережную.

```python
sns.boxplot(x=df['bedrooms'], y=df['price'])

plt.title('Взаимосвязь стоимости дома и Кол-ва спален в доме')
plt.xlabel("Кол-во спален")
plt.ylabel("Цена дома");
```

Можно заметить, что самая высокая цена наблюдается в домах с 8 спальнями.
